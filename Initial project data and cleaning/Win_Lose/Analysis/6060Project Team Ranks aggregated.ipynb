```python
import pandas as pd

# List of file names
file_names = ['Mens Basketball 21-22 NCAA Statistics.xlsx', 'Mens Basketball 20-21 NCAA Statistics.xlsx',
              'Mens Basketball 19-20 NCAA Statistics.xlsx', 'Mens Basketball 18-19 NCAA Statistics.xlsx',
              'Mens Basketball 17-18 NCAA Statistics.xlsx', 'Mens Basketball 16-17 NCAA Statistics.xlsx',
              'Mens Basketball 15-16 NCAA Statistics.xlsx', 'Mens Basketball 14-15 NCAA Statistics.xlsx',
              'Mens Basketball 13-14 NCAA Statistics.xlsx', 'Mens Basketball 12-13 NCAA Statistics.xlsx',
              'Mens Basketball 11-12 NCAA Statistics.xlsx', 'Mens Basketball 10-11 NCAA Statistics.xlsx']

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through each file and read into a DataFrame
for file_name in file_names:
    try:
        # Extract year from filename and prepend "20" to it
        year = "20" + file_name.split(' ')[-3].split('-')[0]

        # Read Excel file into DataFrame
        df = pd.read_excel(file_name)

        # Add a column for the year extracted from the filename
        df['Year'] = year

        # Append the DataFrame to the dfs list
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {file_name}: {e}")

# Concatenate all DataFrames in the list
df = pd.concat(dfs, ignore_index=True)

# Write the merged DataFrame to a new Excel file
df.to_excel("merged_data_MB.xlsx", index=False)

print("Merged data saved to merged_data.xlsx")
```

```python
import statistics

# Read the merged data
df = pd.read_excel("merged_data.xlsx")

# Group by team and calculate mean, median, mode, and standard deviation of rank
team_stats = df.groupby('Team')['Rank'].agg(['mean', 'median', lambda x: statistics.mode(x), 'std']).reset_index()

# Rename the lambda function column to 'mode'
team_stats = team_stats.rename(columns={'<lambda_0>': 'mode'})

# Sort by median rank in ascending order
team_stats = team_stats.sort_values(by='mean')

print("Mens Basketball Statistics for Each Team:")
print(team_stats)
```
